QUE : 1 -->

Train a simple image classifier on the Fashion MNIST dataset using Keras and the SGD optimizer with default parameters, then record the accuracy and loss after 5 epochs.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Normalize pixel values
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Create the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model using SGD optimizer
model.compile(
    optimizer=SGD(),   # Default parameters
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model for 5 epochs
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_data=(x_test, y_test)
)

# Record training and validation results
train_loss = history.history['loss'][-1]
train_accuracy = history.history['accuracy'][-1]
val_loss = history.history['val_loss'][-1]
val_accuracy = history.history['val_accuracy'][-1]

print("\nResults after 5 epochs:")
print("Training Loss:", train_loss)
print("Training Accuracy:", train_accuracy)
print("Validation Loss:", val_loss)
print("Validation Accuracy:", val_accuracy)
```

QUE : 4 -->

Implement a learning rate schedule in your Keras model using the LearningRateScheduler callback to reduce the learning rate by half after every 2 epochs, and train with the Momentum optimizer; report how the learning rate changes during training.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import LearningRateScheduler

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Normalize pixel values
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Build the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Learning rate schedule: halve the learning rate every 2 epochs
def lr_schedule(epoch, lr):
    if epoch > 0 and epoch % 2 == 0:
        return lr * 0.5
    return lr

# LearningRateScheduler callback
lr_scheduler = LearningRateScheduler(lr_schedule, verbose=1)

# Compile the model with SGD + Momentum
model.compile(
    optimizer=SGD(learning_rate=0.01, momentum=0.9),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    x_train,
    y_train,
    epochs=6,
    batch_size=32,
    validation_data=(x_test, y_test),
    callbacks=[lr_scheduler]
)

# Print learning rate after each epoch
lr = 0.01
print("\nLearning Rate Schedule:")
for epoch in range(6):
    if epoch > 0 and epoch % 2 == 0:
        lr *= 0.5
    print(f"Epoch {epoch+1}: Learning Rate = {lr}")

# Print final results
print("\nFinal Training Accuracy:", history.history['accuracy'][-1])
print("Final Validation Accuracy:", history.history['val_accuracy'][-1])
print("Final Training Loss:", history.history['loss'][-1])
print("Final Validation Loss:", history.history['val_loss'][-1])
```

### Why Adam is Often Preferred Over SGD or RMSProp

Adam (Adaptive Moment Estimation) is widely used because it combines the advantages of both Momentum and RMSProp. It automatically adjusts the learning rate for each model parameter, allowing the model to converge faster and more stably than SGD. Compared to RMSProp, Adam generally requires less manual tuning and performs well on a wide range of deep learning tasks, making it a popular choice for real-world applications.

**Real-World Example:**
In a recommendation system like **Flipkart**, the optimizer directly affects how quickly and accurately the model learns user preferences. A well-optimized model using Adam can provide more relevant product recommendations, improving the shopping experience, increasing customer satisfaction, and helping users find products more efficiently.
