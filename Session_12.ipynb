Create a Sequential model in Keras with three layers: an input Dense layer with 32 units and 'relu' activation, a Dense layer with 16 units and 'relu', and an output Dense layer with 1 unit and 'sigmoid'. Print the model summary.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create the Sequential model
model = Sequential([
    Dense(32, activation='relu', input_shape=(10,)),  # Input layer
    Dense(16, activation='relu'),                     # Hidden layer
    Dense(1, activation='sigmoid')                    # Output layer
])

# Print the model summary
model.summary()
```

QUE : 3 -->

Build a simple Keras Functional API model that takes a 10-feature input, passes it through one Dense layer with 20 units and 'relu', then outputs a single value with a Dense layer. Display the model summary and total parameter count.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Define the input layer
inputs = Input(shape=(10,))

# Hidden layer
x = Dense(20, activation='relu')(inputs)

# Output layer
outputs = Dense(1)(x)

# Create the Functional API model
model = Model(inputs=inputs, outputs=outputs)

# Display the model summary
model.summary()

# Display the total number of parameters
print("\nTotal Parameters:", model.count_params())
```

QUE : 4 -->

Write code to check if TensorFlow is using your CPU or GPU for computations. Print out the list of available devices and indicate which one is being used.Hint: Use tf.config.list_physical_devices() and tf.test.is_gpu_available().

```python
import tensorflow as tf

# Print TensorFlow version
print("TensorFlow Version:", tf.__version__)

# List available physical devices
print("\nAvailable Physical Devices:")
print(tf.config.list_physical_devices())

# List CPUs
cpus = tf.config.list_physical_devices('CPU')
print("\nCPU Devices:")
print(cpus)

# List GPUs
gpus = tf.config.list_physical_devices('GPU')
print("\nGPU Devices:")
print(gpus)

# Check whether GPU is available
if len(gpus) > 0:
    print("\nTensorFlow is using GPU for computations.")
else:
    print("\nTensorFlow is using CPU for computations.")
```

QUE : 5 -->

Use ChatGPT or Copilot to generate Keras code for a model with two inputs (e.g., one for user age and one for number of songs listened on Spotify) and one output (predicting likelihood of buying a premium subscription). Paste the generated code, run it, and note any errors or changes you had to make.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model
import numpy as np

# Input 1: User Age
age_input = Input(shape=(1,), name="age_input")

# Input 2: Number of Songs Listened
songs_input = Input(shape=(1,), name="songs_input")

# Combine the inputs
merged = Concatenate()([age_input, songs_input])

# Hidden layers
x = Dense(16, activation="relu")(merged)
x = Dense(8, activation="relu")(x)

# Output layer
output = Dense(1, activation="sigmoid", name="premium_prediction")(x)

# Create the model
model = Model(inputs=[age_input, songs_input], outputs=output)

# Compile the model
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# Display model summary
model.summary()

# Dummy data
age = np.array([[18], [25], [30], [40], [22]], dtype="float32")
songs = np.array([[50], [120], [200], [300], [80]], dtype="float32")
labels = np.array([[0], [1], [1], [1], [0]], dtype="float32")

# Train the model
model.fit(
    [age, songs],
    labels,
    epochs=5,
    batch_size=1
)
```
