Session 21 – Building CNNs in Keras

Task 1–5: code


```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv2D, MaxPooling2D, Flatten,
                                     Dense, BatchNormalization, Dropout)

# Create Sequential Model
model = Sequential()

# Task 1: Conv2D Layer
model.add(Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation='relu',
    input_shape=(32,32,3)
))

# Task 3: Batch Normalization
model.add(BatchNormalization())

# Task 2: Max Pooling
model.add(MaxPooling2D(pool_size=(2,2)))

# Task 2: Flatten
model.add(Flatten())

# Task 3: Dropout
model.add(Dropout(0.3))

# Task 2: Dense Layer
model.add(Dense(64, activation='relu'))

# Output Layer (10 classes for CIFAR-10)
model.add(Dense(10, activation='softmax'))

# Task 4: Compile Model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Show Model Summary
model.summary()

# Generate Random CIFAR-10-like Data
X_train = np.random.rand(100, 32, 32, 3).astype(np.float32)

# Random Labels (One-Hot Encoded)
labels = np.random.randint(0, 10, 100)
y_train = np.eye(10)[labels]

# Train Model
model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=16
)
```

Task 5: Suggested Improvement

Suggestion: Add another Conv2D + BatchNormalization + MaxPooling2D block before flattening. This helps the CNN learn more complex image features and usually improves image classification accuracy.

Improved CNN Code

```python

model = Sequential()

model.add(Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

# Added Improvement
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))

model.add(Flatten())
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```
