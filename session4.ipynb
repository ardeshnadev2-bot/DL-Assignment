```python
#1
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))


def tanh(x):
    return math.tanh(x)


def relu(x):
    if x > 0:
        return x
    else:
        return 0


x = float(input("Enter a number: "))

print("Sigmoid =", sigmoid(x))
print("Tanh =", tanh(x))
print("ReLU =", relu(x))
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 100)


sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0, x)

plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, tanh, label="Tanh")
plt.plot(x, relu, label="ReLU")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()

plt.show()
```

```python
#3
import numpy as np


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)


x = np.arange(-10, 11)

print("x\tSigmoid Derivative")

for i in x:
    print(i, "\t", sigmoid_derivative(i))
```

```python
#4
import numpy as np

def relu(x):
    return np.maximum(0, x)


def dying_relu():
    inputs = np.array([-5, -4, -3, -2, -1])
    outputs = relu(inputs)

    print("Input Array :", inputs)
    print("Output Array:", outputs)

dying_relu()
```

```python
#5
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def modified_sigmoid(x, a):
    return 1 / (1 + np.exp(-a * x))

x = np.linspace(-10, 10, 100)

plt.plot(x, sigmoid(x), label="Original Sigmoid")
plt.plot(x, modified_sigmoid(x, 2), label="Modified Sigmoid (a=2)")

plt.title("Original vs Modified Sigmoid")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()

plt.show()
```
