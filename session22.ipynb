Task 1: Resize 3 Images to 224×224


```python
import os
from PIL import Image

input_folder = "images"
output_folder = "resized_images"

os.makedirs(output_folder, exist_ok=True)

images = ["img1.jpg", "img2.jpg", "img3.jpg"]

for img_name in images:
    img = Image.open(os.path.join(input_folder, img_name))
    img = img.resize((224, 224))

    img.save(os.path.join(output_folder, img_name))

print("All images resized and saved successfully.")
```

Task 2: Convert Color Image → Grayscale → RGB

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("images/img1.jpg")

# Convert BGR to RGB
original = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# RGB to Grayscale
gray = cv2.cvtColor(original, cv2.COLOR_RGB2GRAY)

# Grayscale back to RGB
gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(original)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(gray_rgb)
plt.title("RGB Again")
plt.axis("off")

plt.show()
```

Task 3: Image Augmentation (Rotation + Flip)

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array

img = load_img("images/img1.jpg", target_size=(224,224))
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

datagen = ImageDataGenerator(
    rotation_range=30,
    horizontal_flip=True
)

plt.figure(figsize=(10,10))

i = 1
for batch in datagen.flow(x, batch_size=1):
    plt.subplot(2,2,i)
    plt.imshow(batch[0].astype("uint8"))
    plt.axis("off")
    i += 1

    if i > 4:
        break

plt.show()
```

Task 4: Increase and Decrease Brightness (40%)

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array

img = load_img("images/img1.jpg", target_size=(224,224))
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

# Bright Image
bright = ImageDataGenerator(brightness_range=[1.4,1.4])

# Dark Image
dark = ImageDataGenerator(brightness_range=[0.6,0.6])

bright_img = next(bright.flow(x, batch_size=1))[0].astype("uint8")
dark_img = next(dark.flow(x, batch_size=1))[0].astype("uint8")

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(bright_img)
plt.title("Bright +40%")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(dark_img)
plt.title("Dark -40%")
plt.axis("off")

plt.show()
```

Task 5: Shear Transformation

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array

img = load_img("images/img1.jpg", target_size=(224,224))
x = img_to_array(img)
x = x.reshape((1,) + x.shape)

datagen = ImageDataGenerator(shear_range=25)

sheared = next(datagen.flow(x, batch_size=1))[0].astype("uint8")

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sheared)
plt.title("Sheared")
plt.axis("off")

plt.show()
```
