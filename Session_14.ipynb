QUE : 1 -->

Build a simple movie recommendation neural network using Keras Functional API that takes two inputs: user age and favorite genre (as integers), and outputs a predicted movie rating between 1 and 5.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model

# Input 1: User age
age_input = Input(shape=(1,), name="user_age")

# Input 2: Favorite genre (represented as an integer)
genre_input = Input(shape=(1,), name="favorite_genre")

# Combine the inputs
merged = Concatenate()([age_input, genre_input])

# Hidden layers
x = Dense(16, activation='relu')(merged)
x = Dense(8, activation='relu')(x)

# Output layer (Predicted movie rating between 1 and 5)
output = Dense(1, activation='linear', name="predicted_rating")(x)

# Create the model
model = Model(inputs=[age_input, genre_input], outputs=output)

# Compile the model
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mae']
)

# Display model summary
model.summary()
```

QUE : 2 -->

Create a multi-output neural network using Keras Functional API that predicts both the delivery time (in minutes) and delivery rating (1-5 stars) for a Zomato-style food order, given order amount and distance as inputs.Hint: Use two Dense output layers, one for each prediction.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model

# Input 1: Order Amount
order_amount = Input(shape=(1,), name="order_amount")

# Input 2: Delivery Distance
distance = Input(shape=(1,), name="distance")

# Combine inputs
merged = Concatenate()([order_amount, distance])

# Hidden layers
x = Dense(16, activation='relu')(merged)
x = Dense(8, activation='relu')(x)

# Output 1: Delivery Time (minutes)
delivery_time = Dense(1, activation='linear', name="delivery_time")(x)

# Output 2: Delivery Rating (1–5 stars)
delivery_rating = Dense(1, activation='linear', name="delivery_rating")(x)

# Create the multi-output model
model = Model(
    inputs=[order_amount, distance],
    outputs=[delivery_time, delivery_rating]
)

# Compile the model
model.compile(
    optimizer='adam',
    loss={
        'delivery_time': 'mean_squared_error',
        'delivery_rating': 'mean_squared_error'
    },
    metrics={
        'delivery_time': ['mae'],
        'delivery_rating': ['mae']
    }
)

# Display model summary
model.summary()
```

Modify an existing Keras Sequential model for music playlist recommendations to use the Functional API instead, allowing for future flexibility such as adding more inputs or outputs.Hint: Focus on rewriting the model definition, not the data pipeline.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Input layer
inputs = Input(shape=(10,), name="playlist_features")

# Hidden layers
x = Dense(16, activation='relu')(inputs)
x = Dense(8, activation='relu')(x)

# Output layer
outputs = Dense(1, activation='sigmoid', name="recommendation")(x)

# Create the Functional API model
model = Model(inputs=inputs, outputs=outputs)

# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Display model summary
model.summary()
```

Implement a skip connection in a Keras Functional API model for an IPL cricket score predictor, where the input layer connects both to an intermediate Dense layer and directly to the output layer.Hint: Use the 'Add' layer to combine outputs from different layers.

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Add
from tensorflow.keras.models import Model

# Input layer (e.g., match features)
inputs = Input(shape=(10,), name="match_features")

# Hidden layer
x = Dense(16, activation='relu')(inputs)

# Main branch
main_output = Dense(1, activation='linear')(x)

# Skip connection (input -> output)
skip_output = Dense(1, activation='linear')(inputs)

# Add the main branch and skip connection
final_output = Add(name="predicted_score")([main_output, skip_output])

# Create the Functional API model
model = Model(inputs=inputs, outputs=final_output)

# Compile the model
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mae']
)

# Display model summary
model.summary()
```
