QUE : 1 -->

Create a simple Keras Sequential model for classifying handwritten digits (use the MNIST dataset), and train it without any regularization. Save the validation accuracy after 5 epochs.

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

# Create dummy training data
x_train = np.random.rand(60000, 28, 28).astype("float32")
y_train = np.random.randint(0, 10, 60000)

# Create dummy test data
x_test = np.random.rand(10000, 28, 28).astype("float32")
y_test = np.random.randint(0, 10, 10000)

# Build the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation="relu"),
    Dense(10, activation="softmax")
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Train for 5 epochs
history = model.fit(
    x_train,
    y_train,
    validation_data=(x_test, y_test),
    epochs=5,
    batch_size=32
)

# Print validation accuracy
print("Validation Accuracy after 5 epochs:", history.history["val_accuracy"][-1])
```

QUE : 2 -->

Modify your Keras model to add L2 regularization (kernel_regularizer=regularizers.l2(0.01)) to all Dense layers, retrain for 5 epochs, and compare the validation accuracy to your previous result.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras import regularizers

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Create Sequential model with L2 Regularization
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu',
          kernel_regularizer=regularizers.l2(0.01)),
    Dense(10, activation='softmax',
          kernel_regularizer=regularizers.l2(0.01))
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model for 5 epochs
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    validation_data=(x_test, y_test)
)

# Save validation accuracy after 5 epochs
val_accuracy = history.history['val_accuracy'][-1]

print("\nValidation Accuracy after 5 epochs with L2 Regularization:", val_accuracy)

# Comparison
print("\nCompare this validation accuracy with the previous model (without regularization).")
```

QUE : 3 -->

Add Dropout layers (with rate=0.5) after each Dense layer in your Keras model, retrain for 5 epochs, and observe the change in validation accuracy.Hint: Use keras.layers.Dropout and place it after activations.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Create Sequential model with Dropout
model = Sequential([
    Flatten(input_shape=(28, 28)),
    
    Dense(128, activation='relu'),
    Dropout(0.5),              # Dropout after first Dense layer
    
    Dense(64, activation='relu'),
    Dropout(0.5),              # Dropout after second Dense layer
    
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model for 5 epochs
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_data=(x_test, y_test)
)

# Print validation accuracy
val_accuracy = history.history['val_accuracy'][-1]
print("\nValidation Accuracy after 5 epochs with Dropout:", val_accuracy)

# Compare with previous model
print("Compare this validation accuracy with the model trained without Dropout.")
```

QUE : 4 -->

Integrate BatchNormalization layers into your Keras model (before activation functions in Dense layers), retrain, and note how the training and validation accuracy curves change.Hint: Use keras.layers.BatchNormalization before the activation in each Dense block.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization, Activation

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Create Sequential model with Batch Normalization
model = Sequential([
    Flatten(input_shape=(28, 28)),

    Dense(128),
    BatchNormalization(),
    Activation('relu'),

    Dense(64),
    BatchNormalization(),
    Activation('relu'),

    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model for 5 epochs
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_data=(x_test, y_test)
)

# Print final accuracies
train_accuracy = history.history['accuracy'][-1]
val_accuracy = history.history['val_accuracy'][-1]

print("\nTraining Accuracy after 5 epochs:", train_accuracy)
print("Validation Accuracy after 5 epochs:", val_accuracy)
```

QUE : 5 -->

Implement EarlyStopping in your model training using keras.callbacks.EarlyStopping with patience=2, and show how it affects the number of training epochs and final validation accuracy.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.callbacks import EarlyStopping

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Create Sequential model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# EarlyStopping callback
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=2,
    restore_best_weights=True
)

# Train the model
history = model.fit(
    x_train,
    y_train,
    epochs=20,          # Maximum epochs
    batch_size=32,
    validation_data=(x_test, y_test),
    callbacks=[early_stop]
)

# Print results
print("\nTraining stopped after", len(history.history['loss']), "epochs")
print("Final Validation Accuracy:", history.history['val_accuracy'][-1])
```
