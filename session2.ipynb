```python
#1
import numpy as np

length = int(input("Enter review length: "))
marks = int(input("Enter number of exclamation marks: "))

inputs = np.array([length, marks])

weights = np.array([1, 5])
bias = -50

total = np.dot(inputs, weights) + bias

if total >= 0:
    print("Positive Review")
else:
    print("Negative Review")
```

```python
#2
import numpy as np

x1 = int(input("Enter first input: "))
x2 = int(input("Enter second input: "))

w1 = int(input("Enter first weight: "))
w2 = int(input("Enter second weight: "))

bias = int(input("Enter bias: "))

inputs = np.array([x1, x2])
weights = np.array([w1, w2])

total = np.dot(inputs, weights) + bias

print("Weighted Sum =", total)

if total >= 0:
    print("Output = 1")
else:
    print("Output = 0")
```

```python
#3
import numpy as np


x1 = np.array([[0,0],
               [0,1],
               [1,0],
               [1,1]])


x2 = np.array([0,1,1,0])

weights = np.array([1,1])
bias = -1

for i in range(len(x1)):
    total = np.dot(x1[i], weights) + bias

    if total >= 0:
        prediction = 1
    else:
        prediction = 0

    print("Input:", x1[i], "Prediction:", prediction, "Expected:", x2[i])

print("Note: A single-layer perceptron cannot correctly solve XOR.")
```

```python
#4
import numpy as np


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X = np.array([[0, 0],[0, 1],[1, 0],[1, 1]])

print("XOR using MLP\n")

for x in X:

   
    h1 = sigmoid(x[0] + x[1] - 0.5)
    h2 = sigmoid(x[0] + x[1] - 1.5)

    
    output = sigmoid(20 * h1 - 20 * h2 - 10)

    if output >= 0.5:
        prediction = 1
    else:
        prediction = 0

    print("Input:", x, "Prediction:", prediction)
```

```python
#5
import numpy as np


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


X = np.array([[4.8, 200],[4.5, 150],[3.2, 40],[3.8, 60]])

print("Zomato Restaurant Prediction\n")

for x in X:

 
    rating = x[0] / 5
    reviews = x[1] / 200

    h1 = sigmoid(rating + reviews)
    h2 = sigmoid(rating - reviews)

  
    output = sigmoid(h1 + h2 - 1)

    if output >= 0.5:
        prediction = 1
    else:
        prediction = 0

    print("Rating:", x[0],
          "Reviews:", x[1],
          "Prediction:", prediction)
```
