Save your trained music genre classification model (for example, a model that predicts if a song is Pop, Rock, or Classical based on audio features) to a file using joblib or pickle.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import joblib
import numpy as np

# Create a dummy dataset for demonstration if it doesn't exist
try:
    data = pd.read_csv("music_genre_dataset.csv")
except FileNotFoundError:
    print("Creating a dummy 'music_genre_dataset.csv' file...")
    # Generate some synthetic data
    np.random.seed(42)
    num_samples = 100
    features = pd.DataFrame({
        'feature_1': np.random.rand(num_samples) * 100,
        'feature_2': np.random.rand(num_samples) * 50,
        'feature_3': np.random.rand(num_samples) * 20
    })
    genres = ['Pop', 'Rock', 'Classical']
    labels = np.random.choice(genres, num_samples)

    data = features
    data['genre'] = labels
    data.to_csv("music_genre_dataset.csv", index=False)
    print("'music_genre_dataset.csv' created successfully.")

X = data.drop("genre", axis=1)
y = data["genre"]

# Pipeline
model_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("classifier", RandomForestClassifier(n_estimators=100))
])

# Train
model_pipeline.fit(X, y)

# Save model
joblib.dump(model_pipeline, "music_genre_model.pkl")

print("Model saved successfully!")
```

2.
Build an inference pipeline in Python that loads your saved music genre model and predicts the genre for a new set of audio features provided as input.<br><br><em><strong>Hint:</strong> The pipeline should include data preprocessing steps if required, just like during training.</em>

```python
import joblib
import pandas as pd

# Load saved model
model = joblib.load("music_genre_model.pkl")

new_song = pd.DataFrame({
    "feature_1": [75.0],
    "feature_2": [0.65],
    "feature_3": [10.5]
})

prediction = model.predict(new_song)

print("Predicted Genre:", prediction[0])
```

Create a Streamlit upload interface that lets users upload a CSV file containing audio features of multiple songs, then displays the predicted genre for each song in a table.

```python

import streamlit as st
import pandas as pd
import joblib

model = joblib.load("music_genre_model.pkl")

st.title("🎵 Music Genre Classifier")

uploaded_file = st.file_uploader(
    "Upload CSV file",
    type=["csv"]
)

if uploaded_file is not None:

    data = pd.read_csv(uploaded_file)

    predictions = model.predict(data)

    data["Predicted Genre"] = predictions

    st.subheader("Prediction Results")

    st.dataframe(data)
```
