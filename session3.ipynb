```python
#1


def neuron_output(input_value, weight, bias):
    return input_value * weight + bias

input_value = int(input("Enter input value: "))
weight = int(input("Enter weight: "))
bias = int(input("Enter bias: "))

output = neuron_output(input_value, weight, bias)

print("Neuron Output =", output)
```

```python
#2


def neuron_output(input_value, weight, bias):
    return input_value * weight + bias


likes = [100, 200, 150, 80, 250]

weight = 0.8
bias = 2

for like in likes:
    output = neuron_output(like, weight, bias)
    print("Likes:", like, "Output:", output)
```

```python
#3
import numpy as np

n = 10


zeros = np.zeros(n)


random_weights = np.random.uniform(-1, 1, n)


xavier = np.random.normal(0, 1/np.sqrt(n), n)

print("Weights with Zeros:")
print(zeros)

print("\nRandom Weights:")
print(random_weights)

print("\nXavier Weights:")
print(xavier)
```

```python
#5


def neuron_output(inputs, weights, bias):
    total = 0

    for i in range(len(inputs)):
        total = total + (inputs[i] * weights[i])

    return total + bias


inputs = [2, 4, 1]
weights = [0.6, 0.3, 0.1]
bias = 1

output = neuron_output(inputs, weights, bias)

print("Neuron Output =", output)
```
