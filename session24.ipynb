#TASK2 : Implement a simple RNN cell from scratch in Python that takes a sequence of 5 numbers (e.g., song play counts per day from Spotify) and outputs the final hidden state after processing the sequence.<br><br><em><strong>Hint:</strong> Use the formula: h_t = tanh(Wx_t + Uh_{t-1} + b), where h_0 = 0.</em>

```python
import numpy as np

# 1. The Sequence: 5 days of Spotify song play counts
# We scale them down (divide by 100) so the numbers don't immediately saturate 
# the tanh activation function (which expects smaller numbers).
play_counts = [120, 150, 135, 160, 170]
x = np.array(play_counts) / 100.0 

# 2. Initialize Weights and Bias
np.random.seed(42) # Set a seed so we get the same "random" numbers every time
W = np.random.randn() # Weight for the current input (x_t)
U = np.random.randn() # Weight for the previous hidden state (h_{t-1})
b = np.random.randn() # Bias term

# 3. Initialize the initial hidden state (h_0 = 0) as per instructions
h_prev = 0.0

print(f"Initial Weights -> W: {W:.4f}, U: {U:.4f}, b: {b:.4f}\n")
print("--- Processing Sequence Step-by-Step ---")

# 4. Process the sequence using the RNN formula
for t in range(len(x)):
    x_t = x[t]
    
    # The Core RNN Formula: h_t = tanh(W*x_t + U*h_{t-1} + b)
    z = (W * x_t) + (U * h_prev) + b
    h_t = np.tanh(z)
    
    print(f"Day {t+1} | Input (Plays): {play_counts[t]:3d} | New Hidden State (h_{t+1}): {h_t:.4f}")
    
    # Update the previous hidden state for the next time step
    h_prev = h_t

print("\n--- Final Output ---")
print(f"The final hidden state after processing the whole sequence is: {h_prev:.4f}")
print("This single number now contains a mathematical 'memory' summary of all 5 days!")
```

#Task3:Given the following sentence: 'I love cricket and IPL matches', convert each word to a unique integer (word-to-index mapping), then simulate how an RNN would process this sequence by updating and printing the hidden state at each step.

```python
import numpy as np

# 1. The Sequence: A text sentence
sentence = "I love cricket and IPL matches"
words = sentence.split()

# 2. Word-to-Index Mapping (Tokenization)
# We assign a unique integer to every word in our "vocabulary"

# --- SIMPLER WAY TO DO THIS ---
word_to_idx = {}
current_number = 1

for word in words:
    # If we haven't seen this word yet, give it a new number
    if word not in word_to_idx:
        word_to_idx[word] = current_number
        current_number = current_number + 1

print("--- Step 1: Word-to-Index Mapping ---")
for word in word_to_idx:
    print(f"'{word}' -> {word_to_idx[word]}")

# Convert the whole sentence into a list of numbers
numeric_sequence = []
for word in words:
    number = word_to_idx[word]
    numeric_sequence.append(number)

print(f"\nSentence as numbers: {numeric_sequence}\n")


# 3. Initialize RNN Weights and Bias
np.random.seed(42)
# We multiply by 0.1 to keep weights small. If we multiply an input like '6' 
# by a large weight, the tanh function will max out at 1.0 immediately.
W = np.random.randn() * 0.1  # Input weight
U = np.random.randn() * 0.1  # Hidden state memory weight
b = np.random.randn() * 0.1  # Bias

# Initial hidden state (memory is empty before reading the first word)
h_prev = 0.0

print("--- Step 2: Processing the Sequence ---")

# 4. Simulate the RNN reading the sentence
for t in range(len(numeric_sequence)):
    word = words[t]
    x_t = numeric_sequence[t]
    
    # Calculate the new hidden state using the RNN formula
    z = (W * x_t) + (U * h_prev) + b
    h_t = np.tanh(z)
    
    # Print the current state of the RNN
    print(f"Step {t+1} | Word: '{word:<7}' | Input: {x_t} | New Memory (h_{t+1}): {h_t:.4f}")
    
    # Pass the memory forward to the next time step
    h_prev = h_t

print("\n--- Final Output ---")
print(f"Final hidden state representing the whole sentence: {h_prev:.4f}")
```

Task 4: The Vanishing Gradient Problem

Explanation in Simple Words: 
When an RNN tries to learn, it has to look backward in time to figure out which previous inputs caused an error. To do this, it multiplies the "error gradient" (a number) by its weights over and over again for every step backward in time.Because the weights are usually small numbers (less than 1), multiplying a fraction by a fraction repeatedly makes the number shrink incredibly fast (e.g., $0.5 \times 0.5 \times 0.5 = 0.125$). Eventually, the number becomes so close to zero that it "vanishes." When the gradient vanishes, the network completely stops learning from data that happened many steps ago. It gets "amnesia" for long-term memory.
Real-World Example:
Imagine a text translation app trying to process a very long paragraph:"The boy, who was wearing a red shirt, blue jeans, and carrying a very large backpack filled with heavy textbooks, walked into the room and dropped his bag."If a standard RNN reaches the word "dropped," it needs to predict the word "his" (referring to the "boy"). However, because "boy" was mentioned 20 words ago, the vanishing gradient problem causes the RNN to completely forget the gender of the subject. It might wrongly predict "her" or "their" because its memory has faded.

```python
# 1. Setup Simple Data and Weights
x = [1.0, 2.0, 3.0]  # Our 3-step sequence
target_prediction = 5.0  # What we want the network to predict at the end

W = 0.5  # Weight for the input
U = 0.8  # Weight for the hidden state (memory)
h0 = 0.0 # Initial memory is zero

# ==========================================
# PART 1: THE FORWARD PASS
# ==========================================
print("--- FORWARD PASS ---")
# Step 1
h1 = (W * x[0]) + (U * h0)
print(f"Step 1 Memory (h1): {h1}")

# Step 2
h2 = (W * x[1]) + (U * h1)
print(f"Step 2 Memory (h2): {h2}")

# Step 3 (This is our final prediction)
h3 = (W * x[2]) + (U * h2)
print(f"Step 3 Memory (h3) / Prediction: {h3}")

# ==========================================
# PART 2: CALCULATING THE ERROR
# ==========================================
# How far off was our final prediction from the target?
error = h3 - target_prediction
print(f"\nFinal Error: {error:.2f}")

# ==========================================
# PART 3: BACKPROPAGATION THROUGH TIME (BPTT)
# ==========================================
print("\n--- BACKWARD PASS (BPTT) ---")
# We want to find out how much we need to adjust our memory weight 'U'.
# To do this, the error has to travel backwards through time!

# 1. Gradient at Step 3 (Immediate impact)
grad_step3 = error * h2
print(f"Gradient at Step 3: {grad_step3:.2f}")

# 2. Gradient at Step 2 (Error travels back one step, multiplying by U)
grad_step2 = error * U * h1
print(f"Gradient at Step 2: {grad_step2:.2f}")

# 3. Gradient at Step 1 (Error travels back two steps, multiplying by U * U)
grad_step1 = error * U * U * h0
print(f"Gradient at Step 1: {grad_step1:.2f}")

# The total adjustment for weight 'U' is the sum of gradients from all time steps
total_gradient_U = grad_step3 + grad_step2 + grad_step1
print(f"\nTotal Gradient to update U: {total_gradient_U:.2f}")
```

 brief description of what each part does: 
 
The Forward Pass: The network reads the sequence [1.0, 2.0, 3.0] one step at a time. It updates its hidden state (h1, h2, h3) by combining the current input with the previous step's memory using weights W and U.

Calculating the Error: At the very end of the 3 steps, the model outputs a final prediction. We subtract our target value (5.0) from this prediction to find out how wrong the model was (the error).

Backpropagation Through Time (BPTT): To figure out how to update the memory weight U, the final error has to be carried backward.

At Step 3, the calculation is direct.

At Step 2, the error has to be multiplied by U to jump back in time.

At Step 1, it has to be multiplied by U * U to jump back two steps.

(Note: You can clearly see here that if U was a tiny fraction like 0.1, multiplying by 0.1 * 0.1 at Step 1 would make the gradient vanish, proving the concept from Task 4!)
