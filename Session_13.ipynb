QUE : 4 -->

Simulate a small dataset: generate 100 random user profiles (features: listening_time, liked_genre_score) and a binary label (liked_song: 0 or 1). Train your model using model.fit() for 20 epochs and print the accuracy after training.Hint: Use numpy to generate random data for features and labels.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Generate random dataset
np.random.seed(42)

# Features: listening_time and liked_genre_score
X = np.random.rand(100, 2)

# Binary labels (0 or 1)
y = np.random.randint(0, 2, 100)

# Create the Sequential model
model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X,
    y,
    epochs=20,
    batch_size=10,
    verbose=1
)

# Print the final training accuracy
print("\nTraining Accuracy:", history.history['accuracy'][-1])
```

QUE : 5 -->

Add a callback to your fit() call that stops training early if the validation loss does not improve for 3 epochs.Hint: Use Keras' EarlyStopping callback and set patience=3.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

# Generate random dataset
np.random.seed(42)

# Features: listening_time and liked_genre_score
X = np.random.rand(100, 2)

# Binary labels (0 or 1)
y = np.random.randint(0, 2, 100)

# Create the Sequential model
model = Sequential([
    Dense(16, activation='relu', input_shape=(2,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Create EarlyStopping callback
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

# Train the model
history = model.fit(
    X,
    y,
    epochs=20,
    batch_size=10,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Print final accuracy
print("\nTraining Accuracy:", history.history['accuracy'][-1])
print("Validation Accuracy:", history.history['val_accuracy'][-1])
print("Training stopped after", len(history.history['loss']), "epochs")
```
